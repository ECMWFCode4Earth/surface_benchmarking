# Visualization of locations of the ICOS stations

```r
library(maps)

#directory of icos metadata
dir="/home/lauracma/Documents/ecmwf_proj/data/ISMN_data/in_situ_data/pre_processed_data/station_info/ICOS_FLUXNET/"
files=list.files(dir,full.name=T)

#metadata file 2020
file="/home/lauracma/Documents/ecmwf_proj/data/ISMN_data/in_situ_data/pre_processed_data/station_info/ICOS_FLUXNET/station_coords_2020"
dat=read.table(file,header=T)

dat
```

```r
#count stations per year
nf=length(files)
nstations=array(NA,dim=nf)

for (i in 1:nf) {
    tmp=read.table(files[i],header=T)
    nstations[i]=dim(tmp)[1]
}
```

```r
#par(mfrow=c(1,2))
plot(dat$lon,dat$lat,col=2,lwd=2,pch=3,cex=0.8,xlab="lon",ylab="lat",main="ICOS stations Europe",xlim=c(-16,36),ylim=c(34,70))
map(add=T,col="gray",lwd=2)

years=1996:2020
plot(years,nstations,type="l",xlab="year",ylab="#(available stations)",main="Number of available stations",ylim=c(0,71))
points(years[1],nstations[1],col=2,lwd=2)
points(years[length(years)],nstations[length(years)],col=2,lwd=2)
text(1996,9,"5",lwd=2,col=2)
text(2020,67,"71",lwd=2,col=2)
```
